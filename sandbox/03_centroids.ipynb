```python
import sys
import os
import time
import numpy as np
import numpy.fft as fourier
import math
import time

from gurobipy import *
from matplotlib import colors

import matplotlib.pyplot as plt
import matplotlib as mpl

#import matplotlib.image as mpimg
%matplotlib inline

cmap = colors.ListedColormap(['white', 'green','red', 'violet',"blue"])
bounds=[0,1,2,3,4]
norm = colors.BoundaryNorm(bounds, cmap.N)

def plot_comparison(img0, title0, img1, title1):
    fig2 = plt.figure(figsize = (20, 20))

    sub0 = fig2.add_subplot(2,2,1)
    img_m0 = sub0.imshow(np.flipud(np.transpose(img0)))
    sub0.set_title(title0)
    
    sub1 = fig2.add_subplot(2,2,2)
    sub1.imshow(np.flipud(np.transpose(img1)))
    sub1.set_title(title1)
```

```python
file0 = "./centroid_csv/clean.csv"
file1 = "./centroid_csv/peak.csv"

comp0 = np.genfromtxt(file0, delimiter=',')
comp1 = np.genfromtxt(file1, delimiter=',') 
plot_comparison(comp0, "clean", comp1, "deconvolution")
```
